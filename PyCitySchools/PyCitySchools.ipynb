```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data)
student_data_df = pd.read_csv(student_data)

# Combine the data into a single dataset.  
merge_df = pd.merge(school_data_df, student_data_df, on=["school_name", "school_name"], how="outer")
merge_df
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
total_schools = len(merge_df["school_name"].unique())
```

```python
#Calculate the total number of students
total_students = school_data_df["size"].sum()
formatted_total_students = '{:,}'.format(total_students)
```

```python
#Calculate the total budget
total_budget = school_data_df["budget"].sum()
formatted_total_budget = '${:,.2f}'.format(total_budget)
```

```python
#Calculate the average math score
average_math = round(merge_df["math_score"].mean(), 6)
```

```python
#Calculate the average reading score
average_reading = round(merge_df["reading_score"].mean(), 5)
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
students_math_pass = (merge_df["math_score"] >= 70).sum()
percent_math_pass = round(students_math_pass / total_students * 100, 6)
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
students_read_pass = (merge_df["reading_score"] >= 70).sum()
percent_read_pass = round(students_read_pass / total_students * 100, 6)
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)
both_pass = merge_df[(merge_df["math_score"] >= 70) & (merge_df["reading_score"] >= 70)]["student_name"].count()
percent_both_pass = round(both_pass / total_students * 100, 6)
```

```python
#Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting
district_summary_df = pd.DataFrame({
    "Total School" : [total_schools], 
    "Total Students" : [formatted_total_students],
    "Total Budget" : [formatted_total_budget],
    "Average Math Score" : [average_math],
    "Average Reading Score" : [average_reading],
    "% Passing Math" : [percent_math_pass],
    "% Passing Reading" : [percent_read_pass],
    "% Overall Passing" : [percent_both_pass]})

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
grouped_merge_df = merge_df.groupby(["school_name"])
print(grouped_merge_df)
```

```python
# Obtain School type info
school_type = grouped_merge_df["type"].first() 
```

```python
#Obtain Total Students number for each school
total_student = grouped_merge_df.size()
```

```python
#Obtain Total School Budget
total_budget = grouped_merge_df["budget"].first()
formatted_total_budget = total_budget.map('${:,.2f}'.format)
```

```python
# Calculate Per Student Budget for each school
per_student_budget = total_budget / total_student
formatted_per_student_budget = per_student_budget.map('${:,.2f}'.format)
```

```python
# Calculate Average Math Score for each school
average_math = grouped_merge_df["math_score"].mean()
```

```python
#Average Reading Score for each school
average_reading = grouped_merge_df["reading_score"].mean()
```

```python
# Calculate % Passing Math for each school
students_math_pass = merge_df[merge_df["math_score"] >= 70].groupby(["school_name"]).size()
percent_students_math = students_math_pass / total_student * 100
```

```python
# Calculate % Passing Reading for each school
students_reading_pass = merge_df[merge_df["reading_score"] >= 70].groupby(["school_name"]).size()
percent_students_reading = students_reading_pass / total_student * 100
```

```python
# Calculate % Overall Passing (The percentage of students that passed math and reading) for each school
both_pass = merge_df[(merge_df["math_score"] >= 70) & (merge_df["reading_score"] >= 70)].groupby(["school_name"]).size()
percent_both_pass = both_pass / total_student * 100
```

```python
#Create a dataframe to hold the above results
#Optional: give the displayed data cleaner formatting
grouped_district_summary_df = pd.DataFrame({
    "Total Type" : school_type, 
    "Total Students" : total_student,
    "Total School Budget" : formatted_total_budget,
    "PerStudent Budget" : formatted_per_student_budget,
    "Average Math Score" : average_math,
    "Average Reading Score" : average_reading,
    "% Passing Math" : percent_students_math,
    "% Passing Reading" : percent_students_reading,
    "% Overall Passing" : percent_both_pass})

grouped_district_summary_df.index.name=None
grouped_district_summary_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Make a new dataframe with raw data
school_spending_df = pd.DataFrame({
    "PerStudent Budget" : per_student_budget,
    "Average Math Score" : average_math,
    "Average Reading Score" : average_reading,
    "% Passing Math" : percent_students_math,
    "% Passing Reading" : percent_students_reading,
    "% Overall Passing" : percent_both_pass})
school_spending_df
```

```python
#Create bins and names
spending_bins = [0, 584.99999, 629.9999, 644.9999, 675]
group_names = ["<$585", "$585-630", "$630-645", "$645-675"]
```

```python
#Add a new columns named Spending Ranges Per student and binning based off budget per student
school_spending = school_spending_df.groupby(pd.cut(school_spending_df["PerStudent Budget"], spending_bins, labels=group_names).mean()

school_spending.head(5)
```

```python
#Create a group based off of the bins
scores_spending = scores_spending.groupby("Spending Ranges (Per Student)").mean()
scores_spending.head()
```
